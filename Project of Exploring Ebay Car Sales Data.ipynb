# Which category of majors have the most students?


```python
students_per_cat = {}
categories = recent_grads['Major_category'].unique()

for cat in categories:
    only_cat = recent_grads.loc[recent_grads['Major_category'] == cat]
    total = only_cat["Total"].sum()
    students_per_cat[cat] = total
    
students_per_cat
```

```python
students_cat= pd.Series(students_per_cat)
students_cat = pd.DataFrame(students_cat, columns=['students_cat'])
students_cat.describe()
```

# Revisiting that guy vs girl majors

This time let's use that Major_category column for some variety.

```python
major_category = recent_grads["Major_category"].unique()

women_category = {}
men_category = {}

for c in major_category:
    women_sum = recent_grads.loc[recent_grads["Major_category"] == c, "Women"].sum()
    men_sum = recent_grads.loc[recent_grads["Major_category"] == c, "Men"].sum()
    men_category[c] = men_sum
    women_category[c] = women_sum

women_series = pd.Series(women_category)
men_series = pd.Series(men_category)

men_dataframe= pd.DataFrame(men_series, columns=['Men_categroy'])
women_dataframe= pd.DataFrame(women_series, columns=['Women_categroy'])

major_categories= men_dataframe
major_categories['Women_categroies']=women_dataframe
major_categories
        
```

```python
major_categories[['Men_categroy','Women_categroies']].plot.barh(title='Gap between Women and Men')
```
